```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('disney.csv')
```

```python
# Plotting top categories
top_categories = df['listed_in'].value_counts().head(5)
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.title('Top 5 des meilleures catégories')
plt.xlabel('Catégorie')
plt.ylabel('Nombre de titres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plotting top directors
top_directors = df['director'].value_counts().head(5)
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar', color='salmon')
plt.title('Top 5 des réalisateurs')
plt.xlabel('Réalisateur')
plt.ylabel('Nombre de titres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Les dix meilleurs acteurs
# Drop rows with missing values in the 'cast' column
df_cleaned = df.dropna(subset=['cast'])

# Splitting the cast column into individual actors
actors = df_cleaned['cast'].str.split(', ')

# Creating a list of all actors
all_actors = [actor for sublist in actors for actor in sublist]

# Counting the occurrences of each actor
actor_counts = pd.Series(all_actors).value_counts()

# Getting the top 10 actors
top_actors = actor_counts.head(10)

plt.figure(figsize=(10, 6))
top_actors.plot(kind='bar', color='salmon')
plt.title('Top 10 meilleurs acteurs')
plt.xlabel('Acteur')
plt.ylabel('Nombre de titres')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Les cinq meilleures séries télévisées avec le plus grand nombre de saisons

# Extracting the number of seasons from the 'duration' column and converting it to numeric type
df['num_seasons'] = df['duration'].str.extract('(\d+)').astype(float)

# Filtering TV shows
tv_series = df[df['type'] == 'TV Show']

# Sorting by the number of seasons and selecting the top 5
top_tv_series = tv_series.sort_values(by='num_seasons', ascending=False).head(5)

# Plotting the top TV series
plt.figure(figsize=(10, 6))
plt.bar(top_tv_series['title'], top_tv_series['num_seasons'], color='skyblue')
plt.xlabel('TV Show')
plt.ylabel('Number of Seasons')
plt.title('Top 5 TV Shows with the Highest Number of Seasons')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting movie vs TV show count
disney_movies = df[df['type'] == 'Movie']
disney_tv_shows = df[df['type'] == 'TV Show']
disney_movie_count = len(disney_movies)
disney_tv_show_count = len(disney_tv_shows)
plt.figure(figsize=(6, 6))
plt.bar(['Movies', 'TV Shows'], [disney_movie_count, disney_tv_show_count], color=['gold', 'lightgreen'])
plt.title('Répartition des contenus sur Disney+')
plt.xlabel('Type de contenu')
plt.ylabel('Nombre de titres')
plt.tight_layout()
plt.show()
```
